#### Basic Expressions

This notebook will look at basic expression of Polars to data exploration and basic analysis. A number of these functions exist with other tools like pandas and should be straight forward. This short note will focus on the implementation of the ideas.


##### Reading a CSV File

Let's begin by reading a CSV file

```python
import numpy as np
import polars as pl 

data = pl.read_csv('data/employees.csv')
```

#### head()

The head function returns the top values of the variable that contains the dataframe. You can pass in an n argument to specificy how many rows of data to return. In the example below, I set $n=10$.

```python
print(data.head(n = 10))
```

#### tail()

Much like the <inlinecode>head()</inlinecode> method, the <inlinecode>tail()</inlinecode> method returns the last $n$ values in a dataframe.

```python
print(data.tail(n = 10))
```

#### shape()

The shape method returns a tuple of values containing the number of rows and columns in the dataset. This is also printed along with <inlinecode>head()</inlinecode> and <inlinecode>tail()</inlinecode> methods. These can also be achieved through attributes <inlinecode>height</inlinecode> and <inlinecode>width</inlinecode>

```python
data.shape
```

```python
data.height, data.width
```

#### schema

<inlinecode>schema</inlinecode>  is an attribute that can be both set and retrieved when working with a dataset. It returns the datatypes assigned to all of the columns. Schema can be set for type safety or infered from the dataset itself. Infact, polars gives us options to determine the length of the scan of data to determine these types.

```python
data.schema
```

#### glimpse()

If you have use R and it's suite of tidyverse tools, you may know about the glimpse() method, which literally returns a sample of each column together with the column name, datatype and a few observations.

```python
data.glimpse()
```

#### null_count()

To get a sense of the number of null values in the dataset, the method <inlinecode>null_count()</inlinecode> is provided that returns the number of null values across all columns.

```python
print(data.null_count())
```

#### n_unique()

Similarly, we can get the number of unique counts using the <inlinecode>n_unique()</inlinecode> method. 

```python
data.n_unique()
```

#### columns

The <inlinecode>columns</inlinecode> returns the names of the columns of the data.

```python
data.columns
```
